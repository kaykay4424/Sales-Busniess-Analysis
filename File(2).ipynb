```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from matplotlib import pyplot as plt 
```

```python
# Complete data
df = pd.read_excel(r"C:\Users\kkris\Desktop\Sales_target_1.xlsx")
df
```

```python
#filterning out 
furniture_target = df[df['Category'] == 'Furniture']
```

```python
furniture_target = furniture_target.sort_values('Month of Order Date')
```

```python
# month over month percentage change
furniture_target['Month over Month Percent Change'] = furniture_target['Target'].pct_change() * 100
```

```python
#filtered result 
dt = furniture_target[['Month of Order Date', 'Target', 'Month over Month Percent Change']]
```

```python
#managing date and time format 
furniture_target['Month of Order Date'] = pd.to_datetime(furniture_target['Month of Order Date'], format='%b-%y')
```

```python
furniture_target['Month of Order Date'].dt.strftime('%b-%y')
```

```python
dt
```

```python
#plotting line graph
plt.figure(figsize=(10,5))
plt.plot(furniture_target['Month of Order Date'], furniture_target['Month over Month Percent Change'], marker = 'o', linestyle = '-')
plt.title('Furniture Target Sales: Month over Month % Change')
plt.xticks(furniture_target['Month of Order Date'], furniture_target['Month of Order Date'].dt.strftime('%b-%y'), rotation=45)
plt.xlabel('(Year and Month)')
plt.ylabel('% Change in Target Sales')
plt.xticks(rotation = 0)
plt.grid(True)
plt.show()
```

*(Since the first part of the analysis focuses on the Furniture category, the trend and strategy analysis below also pertain to Furniture.)* *

- Examining the month-over-month percentage change in target sales reveals that most months show stable and incremental growth.
  Keeping out threshold at 5% and treating a spike/dip of 5% or more as significant, We find out that April stands out as the month with significant

_________________________________________________________________________________________________________________________________________________________

**Strategies for aligning target expectations with actual performance trends:**
    
- With people buying furniture in early months of year due to winter sales and discounts. Implement discounts or combo offers in historically 
  weak months (like April) to maintain consistent sales momentum and reduce volatility.
    
- Furniture sales often follow predictable seasonal patterns, with certain months naturally slower than other, thus adjust future targets 
  realistically for such months.

- Furniture purchases are strongly linked to external factors like home sales and household renovation activity. Hence a small decline in real-estate 
  market can also be the reason for dip.
